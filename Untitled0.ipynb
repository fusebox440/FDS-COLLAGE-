```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

pd.set_option("display.max_columns", 50)
pd.set_option("display.width", 120)
sns.set(context="notebook", style="whitegrid")

DATA_PATH = "https://raw.githubusercontent.com/salemprakash/EDA/main/Data/PhDPublications.csv"
df = pd.read_csv(DATA_PATH)

print(df.shape)
print(df.dtypes)
print(df.head(3))
print(df.tail(3))
print(df.isna().sum())
print(df.duplicated().sum())
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = [c for c in df.columns if c not in numeric_cols]

print(numeric_cols)
print(cat_cols)
print(df[numeric_cols].describe().T)
```

```python
def build_quality_report(data, numeric_cols, cat_cols, label="before"):
    rows = []
    for col in numeric_cols:
        s = data[col]
        q1, q3 = s.quantile(0.25), s.quantile(0.75)
        iqr = q3 - q1
        lower, upper = q1 - 1.5*iqr, q3 + 1.5*iqr
        outlier_iqr = ((s < lower) | (s > upper)).sum()
        mean, std = s.mean(), s.std(ddof=0)
        if std == 0:
            outlier_z = 0
        else:
            z = (s - mean) / std
            outlier_z = (z.abs() > 3).sum()
        rows.append({
            "phase": label,
            "column": col,
            "dtype": str(s.dtype),
            "min": s.min(),
            "max": s.max(),
            "mean": s.mean(),
            "std": s.std(),
            "missing": s.isna().sum(),
            "unique": s.nunique(),
            "iqr_outliers_count": int(outlier_iqr),
            "z>3_outliers_count": int(outlier_z)
        })
    for col in cat_cols:
        s = data[col].astype("string")
        rows.append({
            "phase": label,
            "column": col,
            "dtype": "category/string",
            "min": None,
            "max": None,
            "mean": None,
            "std": None,
            "missing": s.isna().sum(),
            "unique": s.nunique(dropna=True),
            "iqr_outliers_count": None,
            "z>3_outliers_count": None
        })
    return pd.DataFrame(rows)

dq_before = build_quality_report(df, numeric_cols, cat_cols, "before")
dq_before.to_csv("data_quality_report_before.csv", index=False)
```

```python
for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(col)
    plt.show()

plt.figure(figsize=(1.2*len(numeric_cols), 5))
sns.boxplot(data=df[numeric_cols], orient="h")
plt.show()

for col in cat_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(x=col, data=df, order=df[col].value_counts().index)
    plt.title(col)
    plt.show()
```

```python
df_clean = df.copy()

# Strip & lowercase categorical columns
for col in ["gender", "married"]:
    df_clean[col] = df_clean[col].astype(str).str.strip().str.lower()

expected_ranges = {
    "articles": (0, 25),
    "mentor": (0, 90),
    "prestige": (1, 5),
    "kids": (0, 5)
}

for col, (lo, hi) in expected_ranges.items():
    df_clean[col] = df_clean[col].clip(lo, hi)

# Winsorization using IQR
def iqr_winsorize(series):
    q1, q3 = series.quantile(0.25), series.quantile(0.75)
    iqr = q3 - q1
    lower, upper = q1 - 1.5*iqr, q3 + 1.5*iqr
    return series.clip(lower, upper)

for col in numeric_cols:
    df_clean[col] = iqr_winsorize(df_clean[col])

# Binary encodings
binary_maps = {
    "gender": {"male":1,"female":0},
    "married": {"yes":1,"no":0}
}

for col, mapping in binary_maps.items():
    if col in df_clean.columns:
        df_clean[col+"_bin"] = df_clean[col].map(mapping)

# Example binning (prestige)
bins = [0, 2, 3, 5]
labels = ["low", "medium", "high"]
df_clean["prestige_group"] = pd.cut(df_clean["prestige"], bins=bins, labels=labels, include_lowest=True)
```

```python
# Split by gender binary for demonstration
males = df_clean[df_clean['gender_bin'] == 1]['articles']
females = df_clean[df_clean['gender_bin'] == 0]['articles']

t_stat, p_val = stats.ttest_ind(males, females, equal_var=False, nan_policy='omit')
print(f"T-test statistic: {t_stat:.4f}, p-value: {p_val:.4g}")

# Cohen's d for effect size
def cohen_d(x, y):
    nx, ny = len(x), len(y)
    dof = nx + ny - 2
    pooled_std = np.sqrt(((nx - 1)*np.var(x, ddof=1) + (ny - 1)*np.var(y, ddof=1)) / dof)
    return (np.mean(x) - np.mean(y)) / pooled_std

print(f"Cohen's d: {cohen_d(males.dropna(), females.dropna()):.4f}")
```
